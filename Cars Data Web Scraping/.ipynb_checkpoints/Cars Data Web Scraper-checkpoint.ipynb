## Cars45 Web Scraper

### Import Library

```python
from bs4 import BeautifulSoup
import requests

import numpy as np
import pandas as pd
```

```python
# car data
car_data = {
    'car_id': [],
    'description': [],
    'amount': [],
    'region': [],
    'condition': [],
    'mileage': [],
    'fuel_type': [],
    'drive_type': [],
    'maker': [],
    'model': [],
    'year_of_manufacture': [],
    'color': [],
    'condition': [],
    'engine_size': [],
    'selling_condition': [],
    'bought_condition': []
}
```

```python
def get_other_info(id):
    url = f'https://www.cars45.com/{id}'
    res = requests.get(url)
    
    soup = BeautifulSoup(res.content, 'html.parser')
```

```python
# funtion to return all car list in a dictionary|
def get_info(car_listings):
    # loop through all list
    for car in car_listings:
        car_id = car['href']
        
        discription = car.find('p', class_="car-feature__name").get_text().strip()
        amount = car.find('p', class_="car-feature__amount").get_text().strip()
        region = car.find('p', class_="car-feature__region").get_text().strip()
        condition = car.find('span', class_="car-feature__others__item").get_text().strip()
        mileage = car.find('div', class_="car-feature__others").find_next('span').find_next('span').get_text().strip()

        # pass all data into car_data dictionary
        car_data['car_id'].append(car_id)
        car_data['description'].append(description)
        car_data['amount'].append(amount)
        car_data['region'].append(region)
        car_data['condition'].append(condition)
        car_data['mileage'].append(mileage)
    
                
```

```python
def main():
    # loop through the pages
    
    for page in range(1,151):
        res = requests.get(f'https://www.cars45.com/listing?page={page}')
    
        soup = BeautifulSoup(res.content, 'html.parser')
        car_listings = soup.find_all('a', class_='car-feature car-feature--wide-mobile')
    
        # invoke the get_info()
        get_info(car_listings)
    
    return car_data
```

```python
data = main()
```

## Convert Data to  DataFrame

```python
car_df = pd.DataFrame(data)
```

## Data Cleaning

```python
car_df.info()
```

```python
# replace '/' with ''
car_df['car_id'] = car_df['car_id'].str.replace('/','')
```

```python
# replace '₦' with ''
car_df['amount'] = car_df['amount'].str.replace('₦','')
```

```python
# replace ',' with ''
car_df['amount'] = car_df['amount'].str.replace(',','')
```

```python
# replace 'km' with ''
car_df['mileage'] = car_df['mileage'].str.replace('km','')
```
